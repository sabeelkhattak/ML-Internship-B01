```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("titanic_cleaned.csv")
```

```python
import os
os.makedirs("visualizations", exist_ok=True)
```

# Line Plot

```python

plt.plot(df['Age'].sort_values())
plt.title("Age Distribution")
plt.xlabel("Index (sorted by age)")
plt.ylabel("Age")
plt.savefig("visualizations/line_plot.png")
plt.show()
```

# Scatter Plot

```python
plt.scatter(df['Age'], df['Fare'])
plt.title("Age vs Fare")
plt.savefig("visualizations/scatter_plot.png")
plt.show()
```

# Histogram

```python
df['Pclass'].hist()
plt.title("Passenger Class Distribution")
plt.savefig("visualizations/histogram.png")
plt.show()
```

# Bar Chart

```python
df.groupby('Sex')['Survived'].mean().plot(kind='bar')
plt.title("Survival by Gender")
plt.savefig("visualizations/bar_chart.png")
plt.show()
```

# Box Plot

```python
sns.boxplot(x='Pclass', y='Fare', data=df)
plt.savefig("visualizations/box_plot.png")
plt.show()
```

# Violin Plot

```python
sns.violinplot(x='Sex', y='Age', data=df)
plt.savefig("visualizations/violin_plot.png")
plt.show()
```

# Heatmap

```python
corr = df.select_dtypes(include='number').corr()

sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.savefig("visualizations/heatmap.png", dpi=300, bbox_inches="tight")
plt.show()
```

# Pair Plot

```python
sns.pairplot(df[['Age', 'Fare', 'Survived', 'Pclass']])
```
