```python
import numpy as np
```

```python
#Create arrays using np.array(), np.zeros(), np.ones(), np.arange()

arr1 = np.array([1, 2, 3, 4, 5])
arr2 = np.zeros((2, 3))
arr3 = np.ones((3, 3))
arr4 = np.arange(0, 10, 2)

arr1, arr2, arr3, arr4
```

# Reshaping

```python
reshaped = arr4.reshape(5, 1)
reshaped
```

# Slicing

```python
slice1 = arr1[1:4]
slice2 = arr3[:, 1]

slice1, slice2
```

# Mathematical Operations

```python
add = arr1 + 10
multiply = arr1 * 2
dot_product = np.dot([1, 2, 3], [4, 5, 6])

add, multiply, dot_product
```

# Statistical Functions

```python
mean = np.mean(arr1)
median = np.median(arr1)
std = np.std(arr1)
variance = np.var(arr1)

mean, median, std, variance
```

# Broadcasting

```python
broadcasted = arr3 + 5
broadcasted
```
